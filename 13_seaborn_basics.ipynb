```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
TOPICS:
seaborn library basics
    sns.distplot(x, y) - disribution plot with KDE(kernel density estimate) + Histogram
        kde, histogram
        bins
        color
        plt.figure() - scale the o/p plot
    sns.countplot(x, y) - freq flot
        hue
        pelette - predefine color scheme
    sns.scatterplot(x, y) - scatter plot
        size
    sns.boxplot(x, y)      - x is categorical data, y is continuous
    sns.pairplot(df)  - data frame with less number of features

In seaborn plots, x, y labels are automatically taken from data frame/series labels 
"""
```

```python
#dataset
df = pd.read_csv('13_dataset_heart.csv')
df.head()
```

```python
#distribution plot - with KDE + histogram
sns.distplot(df['age'])
plt.show()
```

```python
#distribution plot - with histogram
sns.distplot(df['age'], kde = False)
plt.show()
```

```python
#specify no. of bins
sns.distplot(df['age'], kde = False, bins = 5) # divide age data into 5 bins
plt.show()
```

```python
#specify no. of bins
sns.distplot(df['age'], kde = False, bins = 20, color = 'r')
plt.show()
```

```python
#scale o/p figure
plt.figure(figsize = (4,2))
sns.distplot(df['age'], kde = False, bins = 20, color = 'r')
plt.show()
```

```python
#scale o/p figure
plt.figure(figsize = (2,4))
sns.distplot(df['age'], kde = False, bins = 20, color = 'r')
plt.show()
```

```python
#count plot - frequency plot
plt.figure(figsize = (10, 5))
sns.countplot('age', data = df) #another way - sns.countplot(df['age'])
plt.show()
```

```python
#count plot - hue - split data based on given feature
plt.figure(figsize = (10, 5))
sns.countplot('age', data = df, hue = 'sex') # show individual count for each sex - male/female
plt.show()
```

```python
#count plot - hue - split data based on given feature
plt.figure(figsize = (10, 5))
sns.countplot('age', data = df, hue = 'sex', palette = 'twilight' ) # show individual count for each sex - male/female
plt.show()
```

```python
#dataset
df = pd.read_csv('13_dataset_heart.csv')
df.head()
```

```python
#boxplot - x is categorical data, y is continuous data
sns.boxplot('sex', 'age', data = df) #plot age of a person vs sex
plt.show()
```

```python
#boxplot - with hue
sns.boxplot('sex', 'age', data = df, hue ='target') #plot age of a person vs sex . hue as target(heart attack detected) 
plt.show()
```

```python
#dataset
df = pd.read_csv('13_dataset_heart.csv')
df.head()
```

```python
#scatter plot
#plot cholesterol level vs heart rate
sns.scatterplot('chol', 'trestbps', data = df)
```

```python
#scatter plot
#plot age vs heart rate and hue as sex (male/female). 
sns.scatterplot('chol', 'trestbps', data = df, hue = 'sex')
```

```python
#scatter plot
#Here feature 'age' is elected as size parameter to show proportional weight of the feature value 
sns.scatterplot('chol', 'trestbps', data = df, hue = 'sex', size = df['age'])
```

```python
#scatter plot
#color selected as 'terrain'
sns.scatterplot('chol', 'trestbps', data = df, hue = 'sex', size = df['age'], palette = 'terrain')
```
