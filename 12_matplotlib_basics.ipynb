```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""
TOPICS:
    basic plot - scatter, continuous : Ex- plot(), scatter()
    title, x and y axis labels : ex- title(), xlabel(), ylabel()
    x, y axis limits :ex- xlim(), ylim(0)
    plot styles
        color : ex - color = 'red'
        marker style, marker size : ex - marker = 'o', markerfacecolor = 'blue', markersize = 10
        line style: ex- linestyle = '--', linewidth = '2'
"""
```

```python
#data to plot
df = pd.DataFrame({'rooms' : [1, 1, 2, 2, 2, 3, 3], 'rent' : [10000, 5000, 15000, 25000, 20000, 45000, 50000]})
df
```

```python
#plot x, y series data - continuous
plt.plot(df['rooms'], df['rent'])
plt.show()  # optional satement
```

```python
#plot x, y series data - scatter
plt.scatter(df['rooms'], df['rent'])
plt.show()
```

```python
#Display plot title, x, y axis labels
plt.plot(df['rooms'], df['rent'])
plt.title('Housing rents in mumbai')
plt.xlabel('no. of rooms')
plt.ylabel('rent')
plt.show()
```

```python
#Display plot for required x, y data 
plt.plot(df['rooms'], df['rent'])

plt.xlim(1,3)
plt.ylim(10000, 45000)

plt.title('Housing rents in mumbai')
plt.xlabel('no. of rooms')
plt.ylabel('rent')
plt.show()
```

```python
#Plot styling
```

```python
#color - predefined
plt.plot(df['rooms'], df['rent'],
         color = 'red') 

plt.xlim(1,3)
plt.ylim(5000, 50000)

plt.title('Housing rents in mumbai')
plt.xlabel('no. of rooms')
plt.ylabel('rent')
plt.show()
```

```python
#color -  hex code
plt.plot(df['rooms'], df['rent'],
         color = '#5cb504') 

plt.xlim(1,3)
plt.ylim(5000, 50000)

plt.title('Housing rents in mumbai')
plt.xlabel('no. of rooms')
plt.ylabel('rent')
plt.show()
```

```python
#marker settings
plt.plot(df['rooms'], df['rent'],
         color = 'red', 
         marker = 'o', markerfacecolor = 'blue', markersize = 10 ) 

plt.xlim(1,3)
plt.ylim(5000, 50000)

plt.title('Housing rents in mumbai')
plt.xlabel('no. of rooms')
plt.ylabel('rent')
plt.show()
```

```python
#line settings
plt.plot(df['rooms'], df['rent'], 
         color = 'red', 
         marker = 'o', markersize = 10, markerfacecolor = 'blue', 
         linestyle = '--', linewidth = 3) 

plt.xlim(1,3)
plt.ylim(5000, 50000)

plt.title('Housing rents in mumbai')
plt.xlabel('no. of rooms')
plt.ylabel('rent')
plt.show()
```
